```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

# set up parameters for plotting
plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] = 500
plt.rcParams.update({'font.size': 10})
plt.figure(figsize=(12,6))
sns.color_palette("mako", as_cmap=True)
sns.set(font="Verdana")
sns.set(font_scale =1.3)

# read categorial data from csv
df = pd.read_csv('categorial_data.csv', delimiter=';', dtype = {'year': str, 'platform': str, 'frequency': int})

# compute the bubble size as pixel value from frequency by multiplying with a constant value
df['size'] = df.frequency * 230
df = df.sort_values( by=['year','platform','frequency'] )


# use seaborn scatter plot to mimic bubble plot for categorial data
ax = sns.scatterplot(x="year", y="platform", s=df['size'], edgecolor='steelblue', data=df, palette='Blues', color='w')
ax.set_ylim(-1, 4) 
ax.set_xlim(-1, 11) 
ax.set_ylabel('Platform',fontsize=14)
ax.set_xlabel('Year',fontsize=14)
plt.xticks(rotation=90)
plt.gcf().subplots_adjust(bottom=0.25,left=0.18)

# for all x/y combinations write the frequency value in the center of the respective bubble
for index, row in df.iterrows():
    ax.text(row['year'], row['platform'], row['frequency'], ha='center', va='center', size='small', color='b', weight='semibold')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.tight_layout()

```
